***TARGET COLUMN***

---



We chose SES_category as the target column because it was specifically created to represent the socio-economic status (SES) level of each district in our dataset.


---


The creation process was not random — it followed a structured statistical approach that is widely used in socio-economic research:

1. Variable Selection – We included all numeric variables in the dataset that reflect living standards (housing quality, asset ownership, income levels, deprivation measures, etc.).

2. Standardization – Since the variables were on different scales (% values, counts, binary indicators), we converted them into z-scores so each variable contributed equally.

3. PCA (Principal Component Analysis) – We applied PCA to detect patterns in the data. The first principal component (PC1) captured the largest variation, which represented an underlying “wealth/SES” dimension.

4. Orientation Check – We verified that PC1 was positively correlated with indicators of higher SES (e.g., more pucca houses, higher income, more assets). If it were reversed, we would have flipped its sign.

5. Scaling – PC1 scores were rescaled to a 0–100 range for easier interpretation (0 = lowest SES, 100 = highest SES).

6. Categorization – We divided the scaled scores into three equal-sized groups (tertiles): Low SES, Medium SES, and High SES.

7. Final Output – These categories formed the SES_category column, which we set as the target variable for classification tasks.



---



This approach ensures the target column is:

Data-driven (no arbitrary thresholds)

Objective (based on statistical patterns, not assumptions)

Interpretable (higher values = higher SES)

```python
!file /content/drive/MyDrive/final_with_SES
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/final_with_SES')
print(df.shape)
df.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import SelectFromModel
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

# 1. Load and preprocess data
data = df
# Encode target
le = LabelEncoder()
data['SES_category_enc'] = le.fit_transform(data['SES_category'])

# Drop leakage-prone columns
drop_cols = ['district_name', 'SES_category', 'SES_index']
features = data.drop(columns=[c for c in drop_cols if c in data.columns] + ['SES_category_enc'])
target = data['SES_category_enc']

# Fill missing values
features = features.fillna(features.mean())

# Scale features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Feature selection using RandomForest
rf_sel = RandomForestClassifier(n_estimators=100, random_state=42)
rf_sel.fit(features_scaled, target)
selector = SelectFromModel(rf_sel, prefit=True, threshold='median')
features_selected = selector.transform(features_scaled)

# 2. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    features_selected, target,
    test_size=0.2,
    random_state=42,
    stratify=target
)

# 3. SMOTE balancing (training only)
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

# 4. Helper function for model evaluation
def evaluate_model(model, name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    print("="*60)
    print(f"Model: {name}")
    print(f"Accuracy: {acc:.4f}")
    print(f"Weighted F1 Score: {f1:.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred, target_names=le.classes_))
    return acc, f1

# 5. Individual models
results = {}

results['Logistic Regression'] = evaluate_model(
    LogisticRegression(max_iter=1000, random_state=42), "Logistic Regression"
)
results['Random Forest'] = evaluate_model(
    RandomForestClassifier(n_estimators=200, random_state=42), "Random Forest"
)
results['Gradient Boosting'] = evaluate_model(
    GradientBoostingClassifier(n_estimators=200, random_state=42), "Gradient Boosting"
)
results['KNN'] = evaluate_model(
    KNeighborsClassifier(n_neighbors=5), "K-Nearest Neighbors"
)
results['SVM (RBF)'] = evaluate_model(
    SVC(kernel='rbf', C=1, probability=True, random_state=42), "SVM (RBF)"
)
results['XGBoost'] = evaluate_model(
    XGBClassifier(use_label_encoder=False, eval_metric='mlogloss',
                  n_estimators=300, learning_rate=0.1, max_depth=6,
                  subsample=0.8, colsample_bytree=0.8, random_state=42),
    "XGBoost"
)

# 6. Stacking model with GridSearch
clf_rf = RandomForestClassifier(n_estimators=200)
clf_gb = GradientBoostingClassifier(n_estimators=200)
clf_lr = LogisticRegression(max_iter=1000)

stacked_model = StackingClassifier(
    estimators=[('rf', clf_rf), ('gb', clf_gb)],
    final_estimator=clf_lr,
    cv=5
)

params = {
    'rf__n_estimators': [100, 200],
    'final_estimator__C': [0.1, 1.0]
}
grid = GridSearchCV(stacked_model, params, cv=3, scoring='f1_weighted', n_jobs=-1)
grid.fit(X_train, y_train)
y_pred_stack = grid.predict(X_test)

acc_stack = accuracy_score(y_test, y_pred_stack)
f1_stack = f1_score(y_test, y_pred_stack, average='weighted')

print("="*60)
print("Stacking Model (RF + GB -> LR)")
print(f"Accuracy: {acc_stack:.4f}")
print(f"Weighted F1 Score: {f1_stack:.4f}")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_stack))
print(classification_report(y_test, y_pred_stack, target_names=le.classes_))

results['Stacking (RF+GB->LR)'] = (acc_stack, f1_stack)

# 7. Summary table
results_df = pd.DataFrame(results, index=['Accuracy', 'Weighted F1']).T
print("\nModel Comparison:\n")
print(results_df.sort_values(by='Accuracy', ascending=False))
```
